```python
import xarray as xr
import matplotlib.pyplot as plt
from numpy import *

import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
from cartopy.util import add_cyclic_point
```

```python
file='/shared/obs/sw/era40/e4moda.an.sfc.t85.swvl1.1957-2002.nc'
ds_soil=xr.open_mfdataset(file)
ds_soil=ds_soil['SWVL1']*100
ds_soil
```

```python
ds_soils=ds_soil.groupby('time.season').mean(dim='time')
ds_soils
```

```python
clevs=arange(0,110,10)

# Make the figure larger
fig = plt.figure(figsize=(11,8.5))

# Set the axes using the specified map projection
ax=plt.axes(projection=ccrs.Robinson())

# Add cyclic point to data
data=ds_soils.sel(season='DJF')
# data, lons = add_cyclic_point(data, coord=ds_soils['lon'])


# Make a filled contour plot
cs=ax.contourf(ds_soils['lon'], ds_soils['lat'], data, clevs,
            transform = ccrs.PlateCarree(),cmap='Greens',extend='both')

# Add coastlines
ax.coastlines()

# Add gridlines
ax.gridlines()

# Add colorbar
cbar = plt.colorbar(cs,shrink=0.7,orientation='horizontal',label='Soil Moisture %')

# Add title
plt.title('ECMWF ERA-40 Soil Moisture Seasonal Mean - DJF')
```
